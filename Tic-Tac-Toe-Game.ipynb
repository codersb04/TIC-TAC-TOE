```python
# Function to get the winner of the game
def win(arr,player):
            
    if arr[0] == player and arr[1] == player and arr[2] == player:
            return f"{player} is the winner"
```

```python
# Function to check if the player is wins or not
def validate(matrix, player):
    
    size = len(matrix)
    
    # Check for the row
    for row in matrix:
        if win(row, player):
            return win(row,player)
        
    # Check for the column
    for num in range(size):
        col = []
        for row in matrix:
            col.append(row[num])
        if win(col, player):
            return win(col, player)
        
    # Check for Forward Diagonal
    forward = [matrix[0][0], matrix[1][1],matrix[2][2]]
    if win(forward, player):
        return win(forward, player)
    
    # Check for the Backward Diagonal
    backward = [matrix[0][2], matrix[1][1], matrix[2][0]]
    if win(backward, player):
        return win(backward, player)
```

```python
# Function to print the board
def board(matrix):

    print("-------------------")
    print("| ",matrix[0][0]," | ",matrix[0][1]," | ",matrix[0][2]," |")
    print("-------------------")
    print("| ",matrix[1][0]," | ",matrix[1][1]," | ",matrix[1][2]," |")
    print("-------------------")
    print("| ",matrix[2][0]," | ",matrix[2][1]," | ",matrix[2][2]," |")
    print("-------------------")
```

```python
# FUnction to update the board
def update_board(matrix, pos, value):
    
    # Creating list of all the position in the boards: the list is 2D where the each item contains 2 values 
    # representing rows and columns respectively
    pos_list = [[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]]
    row, col = pos_list[pos] 
    
    matrix[row][col] = value # Update the board with new value
    board(matrix)
```

```python
# HUman vs Human
def tic_tac_toe():
    
    #intialize the board with the position values
    ttt_board = [[0, 1, 2],[3,4,5], [6,7,8]]
    board(ttt_board)
    
    #intialize the list with player names
    player = ['X', 'O']

    for i in range(9):
        
        pos = input(f"Chance for {player[i%2]}, Enter the position:")
        update_board(ttt_board, int(pos), player[i%2]) # Update the board according to the position
        
        if i>=3:
            if validate(ttt_board, player[i%2]): # check whether the validate function is return a value
                print(validate(ttt_board, player[i%2]))
                break
       
    else:
        print("Its a TIE")
    
```

```python
tic_tac_toe()
```

```python
# Human vs Computer
import random
def human_computer():
    
    # Create a list of choices containing the possible position
    choices = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    ttt_board = [[0, 1, 2],[3,4,5], [6,7,8]]
    board(ttt_board)
    player = ['X', 'O']
    
    for i in range(9):
        
        if i%2==0:
            pos = input(f"Chance for {player[i%2]}, Enter the position:")
            choices.remove(int(pos)) #remove the input position from the list
            update_board(ttt_board, int(pos), player[i%2]) # update the board
            
        else:
            print("Chance for Computer: \n")
            comp_pos = random.choice(choices) # randomly choose the position available from the list
            choices.remove(int(comp_pos)) 
            update_board(ttt_board, int(comp_pos), player[i%2])
            
        if i>=3:
            if validate(ttt_board, player[i%2]):
                print(validate(ttt_board, player[i%2]))
                if i%2==0:
                    print("Human Wins")
                else:
                    print("Computer Wins")
                break
       
    else:
        print("Its a TIE")
```

```python
human_computer()
```
